```python
import numpy as np
data = [4, 8, 15, 21, 21, 24, 25, 28, 34]
bin_boyutu = 3
```

```python
#SMOOTHING BY BIN MEANS
```

```python
b1 = np.zeros((int(len(data) / bin_boyutu), bin_boyutu))

for i in range (0, len(data), bin_boyutu):
    k = int(i/bin_boyutu)
    ortalama = sum(data[i: i+bin_boyutu]) / bin_boyutu
    b1[k:bin_boyutu] = ortalama
print("-----------------Ortalama Bin:----------------- \n",b1)
```

```python
#SMOOTHING WITH BOUNDRIES
```

```python
b3 = np.zeros((int(len(data) / bin_boyutu), bin_boyutu))

for i in range (0, len(data), bin_boyutu):
    k = int(i/bin_boyutu)
    for j in range (bin_boyutu):
        if (data[i+j]-data[i]) < (data[i+2]-data[i+j]):
            b3[k,j]=data[i]
        else:
            b3[k,j]=data[i+2]

print("-----------------Sƒ±nƒ±rlayƒ±cƒ± Bin:----------------- \n",b3)
```
